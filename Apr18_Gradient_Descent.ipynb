```python
import numpy as np
from numpy.random import random, randn
from matplotlib import pyplot as plt, animation, rcParams, rc
```

```python
def linear_function(alpha, beta, x): return alpha + beta * x
rc('animation', html='html5')
rcParams['figure.figsize'] = (3,3)
```

```python
beta_true = 6
alpha_true = 3
n = 30
x = random(n)
y = linear_function(alpha_true, beta_true, x) + 0.2*randn(n)
beta = -1
alpha = 2
lr = 0.01
```

```python
def update_alpha_and_beta():
    global alpha, beta
    y_predicted = linear_function(alpha, beta, x)
    dydalpha = 2 * (y_predicted - y)
    dydbeta = x * dydalpha
    alpha = alpha - lr*dydalpha.mean()
    beta = beta - lr*dydbeta.mean()
```

```python
def animate(i):
    x = np.linspace(0, 1, 100)
    y = linear_function(alpha, beta, x)
    line.set_data(x,y)
    for i in range(20):
        update_alpha_and_beta()
    return(line,) 
```

```python
fig = plt.figure(dpi=80, figsize=(7,7))
ax = fig.add_subplot(111)
ax.set_xlim((0,1))
ax.set_ylim((-2,15))
plt.scatter(x,y)
line, = ax.plot([], [], lw=2)
ani = animation.FuncAnimation(fig, animate, np.arange(0, 150), interval = 100)
```
